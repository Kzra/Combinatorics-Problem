**This script computes which integers can be represented by consecutive pairs of smaller integers:** 

All time signatures are simple, complex or irregular, meaning they break down into two or three (or a combination of) 'strong beats'.

No matter what the uppernumerical is in a time signature, it can be described as simple or complex or irregular, because all integers >3 can be represented by groupings of 2 and 3. 
4 = 2\*2, 5 = 2 + 3, 6 = 2/*3, etc.

How does this apply to pairs on integers >3, and this script is to show the pattern that emerges when we use pairs of integers >3 to describe numbers.


```python
import numpy as np
from matplotlib import pyplot as py
%matplotlib inline
```

```python
#This is a function that outputs any numbers up to n + int 2 that can't be created by a combination of two input integers. 
def doesnt_combine (int1,int2,n):
    invalid_ints = []
    for i in range(int2,(n+int2)):
        valid = combine_check(int1,int2,i)
        if valid == False:
            invalid_ints.append(i)
    return invalid_ints

#This is a function that checks whether integers can be made up from other integers 
def combine_check(int1,int2,qint):
    valid = False
    if qint % int1 == 0:
        valid = True
        return valid
    if qint % int2 == 0:
        valid = True
        return valid
    if (qint-int2) % int1 == 0:
        valid = True
        return valid
    if (qint-int1) % int2 == 0:
        valid = True
        return valid
    #A vectorized operation to check whether any valid combinations exist
    m1 = int(qint/int1)
    arr = np.full((1,m1),qint)
    arr2 = np.zeros(m1)
    for i in range(0,m1):
        arr2[i] = int1*(i+1)     
    arr3 = arr - arr2
    varr = arr3 % int2
    vcheck = np.where(varr == 0)
    if len(vcheck[0]) > 0:
        valid = True
        return valid
    return valid
```

```python
no_invals= []
combo = []
for i in range(2,100):
    r = doesnt_combine(i,i+1,10000)
    print(i,',',i+1,': ',len(r))
    print(r)
    no_invals.append(len(r))
    combo.append(i)
    print('\n')
py.plot(combo,no_invals)
py.xlabel('combination (n,n+1) ')
py.ylabel('# ints that cant be described')
```
